```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
# Read the airports.csv and display dataframe
topcities_df = pd.read_csv('topcitiescensusinfo.csv')
topcities_df
```

```python
# Create a scatter plot of Population vs MEDIAN HOUSEHOLD INCOME ($)
plt.scatter(topcities_df["POPULATION"],topcities_df["MEDIAN HOUSEHOLD INCOME ($)"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("City Population vs Median Household Income")
plt.xlabel("Population")
plt.ylabel("Median Household Income ($)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationhouseholdincome.png")
plt.show()
```

```python
# Create a scatter plot of Population vs POVERTY RATE (%)
plt.scatter(topcities_df["POPULATION"],topcities_df["POVERTY RATE (%)"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("City Population vs Poverty Rate")
plt.xlabel("Population")
plt.ylabel("Poverty Rate (%)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationpoverty.png")
plt.show()
```

```python
# Create a scatter plot of Population vs EMPLOYMENT RATE (%)
plt.scatter(topcities_df["POPULATION"],topcities_df["EMPLOYMENT RATE (%)"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("City Population vs Employment Rate")
plt.xlabel("Population")
plt.ylabel("Employment Rate (%)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationemployment.png")
plt.show()
```

```python
# Create a scatter plot of Population vs BACHELOR'S DEGREE OR HIGHER (%)
plt.scatter(topcities_df["POPULATION"],topcities_df["BACHELOR'S DEGREE OR HIGHER (%)"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("City Population vs Bachelor's Degree or Higher")
plt.xlabel("Population")
plt.ylabel("Bachelor's Degree or Higher (%)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationdegree.png")
plt.show()
```

```python
# Create a scatter plot of MEDIAN HOUSEHOLD INCOME ($) vs BACHELOR'S DEGREE OR HIGHER (%)
plt.scatter(topcities_df["MEDIAN HOUSEHOLD INCOME ($)"],topcities_df["TOTAL HOUSING UNITS"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("Median Household Income vs Total Housing Units")
plt.xlabel("Median Household Income")
plt.ylabel("Total Housing Units")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationdegree.png")
plt.show()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(topcities_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, topcities_df["POPULATION"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, topcities_df["CITY"], rotation="vertical")
plt.title("Population by City with Airports with 2,000,000+ passengers")
plt.xlabel("Cities")
plt.ylabel("Population")
plt.grid()
# Save a picture of the plot as .png file
# plt.savefig("Images/citiespopulationbarplot.png")
plt.show()
```

```python
# Linear Regression equation and Print out the r-squared value along with the plot.
x_values = topcities_df["MEDIAN HOUSEHOLD INCOME ($)"]
y_values = topcities_df["TOTAL HOUSING UNITS"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,10),fontsize=15,color="red")
plt.xlabel("Median Household Income ($)")
plt.ylabel("Total Housing Units")
print(f"The r-squared is: {rvalue**2}")
print(f"The line equation is: {line_eq}")
# Save a picture of the plot as .png file
# plt.savefig("Images/hhincomevhousingunits.png")
plt.show()
```

```python
# Rename columns to prepare dataframe for merge
topcities_df.rename(columns={"CITY":"City"}, inplace=True)
topcities_df.head()
```

```python
# Read the airport_ratings_passengers.csv and display dataframe
airport_sales_df = pd.read_csv('airport_sales.csv')
airport_sales_df.head()
```

```python
# Merge topcities_df with airport_sales_df by "City"
merged_df = pd.merge(topcities_df, airport_sales_df, on="City", how="left")
merged_df
```

```python
# merged_df.fillna(value='None',inplace=True)
merged_df.dropna(how='any', inplace=True)
merged_df
```

```python
# Linear Regression equation and Print out the r-squared value along with the plot.
x_values = merged_df["POPULATION"]
y_values = merged_df["Originating Domestic Passengers"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,20),fontsize=15,color="red")
plt.xlabel("Population in cities with airports")
plt.ylabel("Airport passengers (in millions)")
print(f"The r-squared is: {rvalue**2}")
print(f"The line equation is: {line_eq}")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationvpassengers.png")
plt.show()
```
