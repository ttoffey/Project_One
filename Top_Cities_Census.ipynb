```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
# Read the airports.csv and display dataframe
topcities_df = pd.read_csv('topcitiescensusinfo.csv')
topcities_df
```

```python
# Create a scatter plot of Population vs MEDIAN HOUSEHOLD INCOME ($)
plt.scatter(topcities_df["POPULATION"],topcities_df["MEDIAN HOUSEHOLD INCOME ($)"],edgecolors="black",facecolors="lightblue",alpha=1)
plt.title("City Population vs Median Household Income")
plt.xlabel("Population")
plt.ylabel("Median Household Income ($)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="grey")
# Save a picture of the plot as .png file
# plt.savefig("Images/populationhouseholdincome.png")
plt.show()
```
