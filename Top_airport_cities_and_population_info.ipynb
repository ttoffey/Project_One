```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
merged_df=pd.read_csv('./Resource Files/topcitiescensusinfo.csv')
merged_df
```

```python
# Using scatter plot to determine the commonalities between Population and total housing unit
x_values = merged_df["POPULATION"]
y_values = merged_df["TOTAL HOUSING UNITS"]

plt.scatter(x_values,y_values)
plt.title('Commonalities between Aiport Population and Housing Units')
plt.xlabel('POPULATION')
plt.ylabel('TOTAL HOUSING UNITS')
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.tight_layout()
plt.show()
```

```python
# Using scatter plot to determine the commonalities between Population and Employment Rate
x_values = merged_df["POPULATION"]
y_values = merged_df["EMPLOYMENT RATE (%)"]

plt.scatter(x_values,y_values)
plt.title('Commonalities between Aiport Population and Employment Rate')
plt.xlabel('POPULATION')
plt.ylabel('EMPLOYMENT RATE (%)')
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.tight_layout()
plt.show()
```

```python
# Using scatter plot to determine the commonalities between Population and Poverty Rate 
x_values = merged_df["POPULATION"]
y_values = merged_df["POVERTY RATE (%)"]

plt.scatter(x_values,y_values)
plt.title('Commonalities between Aiport Population and Poverty Rate')
plt.xlabel('POPULATION')
plt.ylabel('POVERTY RATE (%)')
plt.minorticks_on()
plt.grid(which='major', linestyle='-')
plt.grid(which='minor', linestyle=':')
plt.tight_layout()
plt.show()
```
